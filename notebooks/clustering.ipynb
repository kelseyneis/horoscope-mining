```python
#import libraries
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# read in data.
df = pd.read_csv('../horoscope_saved.csv')
df = df[df['category']=='general']
df = df.drop(columns=['category', 'date'])

signs = {'aries': 1, 'taurus': 2, 'gemini': 3, 'cancer': 4, 'leo': 5, 'virgo': 6, 'libra': 7, 'scorpio': 8, 'sagittarius': 9, 'capricorn': 10, 'aquarius': 11, 'pisces': 12}

df['cluster'] = df['sign'].map(signs)
y = np.array(df['cluster'])
X = np.array(df['horoscope'])
```

```python
vectorizer = CountVectorizer(stop_words='english')
vectorizer.fit(np.array(X))
X = vectorizer.transform(X)
X = X.toarray()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=50)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents)
```
